#  **1. Data Cleaning**




```python
#Importing Libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#Load_dataset
df = pd.read_csv("FDI_in_India.csv")
```

#  **2. Exploratory Data Analysis**

```python
#Calculate basic summary
df.describe()

#Plotting time series data :
plt.figure(figsize = (12,6))

#Plotting time series data for each sector
for Sector in df.columns[1:]:
    plt.plot(df['Sector'],df[Sector],label=Sector)
plt.xlabel("Year")
plt.ylabel("FDI Value")
plt.title("FDI trends by Sector")
plt.legend()
plt.xticks(rotation=90)
plt.show()



```

```python
#Identify sectors with highest and lowest FDI
highest_fdi_sectors = df.set_index('Sector').sum(axis=1).nlargest(5)
lowest_fdi_sectors = df.set_index('Sector').sum(axis=1).nsmallest(5)
print("Sectors with highest FDI:\n", highest_fdi_sectors)
print("Sectors with lowest FDI:\n", lowest_fdi_sectors)

```

```python
#Correlations with different sectors
from sklearn.preprocessing import StandardScaler

# Scale the FDI data
scaler = StandardScaler()
scaled_data = scaler.fit_transform(df.drop('Sector', axis=1))

# Calculate the correlation matrix for scaled data
correlation_matrix_scaled = pd.DataFrame(scaled_data, columns=df.drop('Sector', axis=1).columns).corr()

# Plot the heatmap with scaled data
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix_scaled, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5,
            xticklabels=df['Sector'], yticklabels=df['Sector'])
plt.title('Correlation Matrix (Scaled Data)')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12, 6))
sns.set_palette("tab20")  # Set the color palette
sns.set_style("whitegrid")  # Set the style

sectors = df.columns[1:]

# Initialize the y_offset to store the stacked values
y_offset = pd.Series(0, index=df.index)

for sector in sectors:
    plt.fill_between(df['Sector'], y_offset, y_offset + df[sector], label=sector)
    y_offset += df[sector]

plt.xlabel('Year')
plt.ylabel('FDI Value')
plt.title('Composition of FDI across Sectors')
plt.legend(loc='upper left', bbox_to_anchor=(1, 1))
plt.xticks(rotation=90)
plt.tight_layout()  # Ensures all labels are visible
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Calculate the percentage growth of each sector over the years
percentage_growth = df.drop('Sector', axis=1).pct_change() * 100

# Replace inf and NaN values with 0
percentage_growth.replace([np.inf, -np.inf, np.nan], 0, inplace=True)

# Calculate the mean percentage growth for each sector and sort them in descending order
average_growth = percentage_growth.mean().sort_values(ascending=False)

# Plot the average growth for each sector
plt.figure(figsize=(12, 6))
sns.barplot(x=average_growth.index, y=average_growth.values, palette='tab20')
plt.xlabel('Sector')
plt.ylabel('Average Percentage Growth')
plt.title('Average Percentage Growth by Sector')
plt.xticks(rotation=90)
plt.tight_layout()  # Ensures all labels are visible
plt.show()

```

#  **5. Sector Contribution**

```python
plt.figure(figsize=(12, 6))
for sector in df.columns[1:]:
    plt.plot(df['Sector'], df[sector], label=sector)
plt.xlabel('Year')
plt.ylabel('FDI Value')
plt.title('FDI Trends by Sector')
plt.legend()
plt.xticks(rotation=90)
plt.show()
```

```python
# Assuming the dataset contains columns 'Sector', '2001-02', ..., '2016-17'
df['FDI_Growth_Rate'] = (df['2016-17'] / df['2001-02'] - 1) * 100
sector_highest_growth = df[df['FDI_Growth_Rate'] == df['FDI_Growth_Rate'].max()]['Sector'].values[0]
print("Sector with highest FDI growth rate between 2001 and 2017:", sector_highest_growth)
```

```python
# Assuming the dataset contains columns 'Sector', '2001-02', ..., '2016-17'
fdi_data_overall = df.drop('Sector', axis=1).mean()
plt.plot(fdi_data_overall.index, fdi_data_overall.values, marker='o')
plt.xlabel('Year')
plt.ylabel('Average FDI Value')
plt.title('Overall FDI Trend in India (2001-2017)')
plt.xticks(rotation=45)
plt.show()

```

```python
# Assuming the dataset contains columns 'Sector', '2001-02', ..., '2016-17'
fdi_data_cyclic = df.drop('Sector', axis=1).T
fdi_data_cyclic.plot(legend=False, alpha=0.5, colormap='viridis')
plt.xlabel('Year')
plt.ylabel('FDI Value')
plt.title('Cyclic FDI Trends in Different Sectors (2001-2017)')
plt.legend(loc='upper left', bbox_to_anchor=(1.05, 1), title='Sector', labels=df['Sector'])
plt.show()
```
